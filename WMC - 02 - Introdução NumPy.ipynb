Os arrays podem ter mais que duas dimensões, sendo também de 3 e 4 dimensões, onde estas últimas são bem difíceis de visualizar mentalmente e são bastante usadas em redes neurais.

Abaixo, temos 3 arrays 2d = 2 arrays de 2 linhas e 2 colunas, que geram um array 3d = 3 arrays de 2 linhas e 2 colunas. Um array 3d é composto de altura, largura e comprimento/profundidade.
Os arrays abaixo estão em formato de lista.

```python
arr1_2d = np.array([[1,2],[3,4]])
arr2_2d = np.array([[5,6],[7,8]])
arr3_2d = np.array([[9,2],[3,4]])

arr_3d = np.array([arr1_2d, arr2_2d, arr3_2d])
arr_3d
```

Arrays de 4 dimensões não são visíveis. Temos o código abaixo, mas sua visualização gráfica não é possível.

```python
arr1 = np.array([arr_3d, arr_3d])
arr1
```

Vetores - Array 3D
Programadores e a documentação do numpy se referem ao array como vetores, tensores ou matrizes. Termos matemáticos.
Os vetores, dentro do numpy, se referem ao array como de uma dimensão só ex: shape(4,) que pode ser uma linha e 4 colunas OU 1 coluna e 4 linhas.
Para criar um array que seja expecificamente vertical ou horizontal, precisamos colocar a ordem de como queremos, linhas e colunas.
Na matemática, podemos considerar coisas 2D como matrizes e 3D como tensores.



Atributos e métodos do array:

Atributos são propriedades qe geram valores e resultados, características do objeto. .shape = mostra quantas linhas e quantas colunas tem o array.

Os métodos geram ações, como achatar array, transformar dimensões etc. .flatten = achata a array, transformando em um array de uma dimensão (como uma lista). .reshape = muda as dimensões do array, uma que teria 2 linhas e 3 colunas, será remodelada para 3 linhas e 2 colunas. Nesse caso, não pode haver mudança no número de elementos, se eu tenho um array de 4 linhas e 3 colunas, isso terá 12 elementos, só posso remodelar para um array com 12 elementos, ex: 6 linhas e 2 colunas.

2 - Considerando que você é uma organizadora de um jogo de bingo. Crie um array que irá representar a cartilha desses jogos de bingo. Os números das suas cartelas variam entre 1 a 30, e você terá 10 participantes. Cada cartela terá 12 números (4,3).
Crie um array que represente esse jogo.

```python
abingo = np.random.randint(1,31, size=(10, 4, 3))
abingo
```

3 - Faça o reshape das suas cartelas para que haja 5 cartelas de 4 linhas e 6 colunas.

```python
abingo = abingo.reshape((5, 4, 6))
abingo
```
