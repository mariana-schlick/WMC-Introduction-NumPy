O 64 bits será usado para 64 bits e o 32 para 32. Isso é feito para economizar memória. Se tenho uma array quen não necessita 64 bits não pe necessário alocar em 64 bits.

Bits unidade binária, menor unidade de dados de um computador, 0 ou 1.

Bytes é uma sequência de 8 Bits.

2024 em binário = 0000011111101000 = 2bytes

Fica armazenado na np.int32 que armazena 4.294.967.296 bits em valor inteiro = 2³²

Dividio metade para mais e metade para menos.

Quando é 64, fica 2 elevado a 64, gerando um número MUITO maior.



Atributo .dtype e tipo default

Esse atributo indica qual é o tipo dos elementos que estão sendo guardados.

```python
np.array([2.14, 6.25, 160.87, 8.4]).dtype
```

```python
np.array([2, 6, 160 , 8]).dtype
```

Dados tipo String são expressados de forma um pouco diferente:

```python
np.array(["Hello", "Coder", "Girls"]).dtype
```

O valor gerado foi o <U5, pois as palavras contém 5 letras.

Se no meio das palavras de 5 letras, tivesse alguma com mais letras, seria considerado esse valor como U

```python
np.array(["Hello", "Coders", "Girls"]).dtype
```

Conversão e coerção

dtype como argumento:

```python
arr_float32 = np.array([2.14, 6.25, 160.87, 8.4], dtype=np.float32)
arr_float32
```

Conversão - Tudo que é false é 0 e tudo que é true é 1, ver tabela verdade.

```python
arr_vf = np.array ([[True, False], [True, True]], dtype=np.bool8)
arr_vf.astype(np.int32)
```

Coerção

```python
np.array(["pedra", False, 42, 42.42])
```

Hierarquia de coerção

Adicionar um float a um array de int mudará todos os valores para float.

```python
np.array([0, 10]).dtype
```

```python
np.array([0, 10, 1.]).dtype
```

Adicionar um inteiro em um array de booleanos transformará todo o array em itneiros

```python
np.array([True, False]).dtype
```

```python
np.array([True, False, 12]).dtype
```
