Concatenação = União de coisas, pode ser listas, arrays, strings etc

Concatenando linhas - adicionar uma linha logo abaixo de outra na array

O Array indicado no inicio fica por cima e logo após os subsequentes.

np.concatenate((arrv1, arrv2))





```python
arrv1 = np.random.randint(10, size=(3,2))
arrv1
```

```python
arrv2 = np.array([["Peras", "Morangos"]])
arrv2
```

```python
np.concatenate((arrv1, arrv2))
```

Concatenação de colunas - usamos o axis=1

```python
arrv1
```

```python
arro1 = np.array(["Uva", "Abacaxi", "Laranja"]).reshape((3, 1))
arro1
```

```python
np.concatenate((arrv1, arro1), axis=1)
```

SE TENTARMOS CONCATENAR LINHAS E COLUNAS EM QUANTIDADES QUE NÃO SÃO EQUIVALENTES AO OUTRO ARRAY, DARÁ ERRO

exemplo, se um array com 3 linhas e 4 colunas eu queira adicionar um array com 2 linhas e 3 colunas. Isso também nos diz que precisamos informar todos os matches, como se tenho um array com 3 linhas e 4 colunas eu queira adicionar um array com 3 linhas e 1 coluna, preciso informar o shape como (4, 1) e não somente (4,)

É possível adicionar arrays para criar novas dimensões, mas o método np.concatenate não permite. Além disso, só é compatível para as mesmas dimensões já existentes no array.

Deletando com axis=0 e com axis=1

```python
sala_de_espera = np.array([['5', '30', '1', 'Alice'],
                       ['6', '29', '1', 'Bob'],
                       ['7', '35', '3', 'Cristina'],
                       ['8', '34', '3', 'Luiz']])
sala_de_espera
```

```python
np.delete(sala_de_espera, 2, axis=0)
```

```python
np.delete(sala_de_espera, 1, axis=1)
```

Cálculos com Arrays

Métodos de agregação de dados

Python - .sum()  |  .min()  |  .max()  |  .mean()  |  .cumsum()

Temos dados de acidentes, onde cada coluna é um cliente e cada linha é um ano.



```python
acidentes = np.array(([1, 3, 2],
                     [0, 1, 0],
                     [2, 1, 4],
                     [0, 0, 0],
                     [1, 1, 0]))
acidentes
```

Se eu quiser somar tudo, uso .sum() que irá somar o array inteiro

```python
acidentes.sum()
```

Se eu quiser somar colunas, uso .sum(axis=0) que irá somar as colunas separadamente

```python
acidentes.sum(axis=0)
```

Se eu quiser somar linhas, uso .sum(axis=1) que irá somar as linhas separadamente

```python
acidentes.sum(axis=1)
```

Mostrar os mínimos e máximos do array


```python
acidentes.min()
```

```python
acidentes.max()
```

Máximo de cada linha

```python
acidentes.max(axis=0)
```

Máximo de cada coluna

```python
acidentes.max(axis=1)
```

Média


```python
acidentes.mean()
```

```python
acidentes.mean(axis=0)
```

```python
acidentes.mean(axis=1)
```

Método keepsims - manter as mesmas dimensões da array

```python
acidentes.mean(axis=0)
```

```python
acidentes.mean(axis=0, keepdims=True)
```

```python
acidentes.mean(axis=1)
```

```python
acidentes.mean(axis=1, keepdims=True)
```

Soma cumulativa - cumcum()

É interessante para fazer plotagem de histórico

```python
acidentes.cumsum(axis=0)
```

```python
acidentes.cumsum(axis=1)
```
