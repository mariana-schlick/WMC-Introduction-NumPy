Criando um numpy array do zero


Nesse primeiro exemplo temos uma array onde você escolhe a dimensão em linhas e colunas, gerando uma array de elementos zeros. Esse np.zeros recebe uma tupla, que são (), a escrita da array já possuí (), recebendo a tupla, ficam duas duplas de (()).

```python
np.zeros((3,2))
```

Aqui temos o exemplo de range, onde você coloca o valor inicial e o valor final e ele dará a ordem em números inteiros. Observe que o 0 está incluído na contagem, mas o 3 não entra. Ou seja, se fosse para o 3 aparecer, teria que ser ((0, 4))

```python
list(range(0,3))
```
