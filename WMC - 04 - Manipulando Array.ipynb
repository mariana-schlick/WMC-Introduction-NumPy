```python
import numpy as np
```

Manipulando Array

Index em um array de 1 dimensão

```python
arr = np.array([2, 4, 6, 8])
arr[2]
```

Index em um array de 2 dimensões

selecionar número específico em uma linha e coluna

```python
arr = np.array([[ 9, 24,  1],
                [13, 10, 20],
                [27, 10,  7],
                [24,  2, 15]])
arr[1, 2]
```

selecionar uma linha inteira

```python
arr[0]
```

selecionar coluna inteira

```python
arr[:, 1]
```

Fatiar (slicing) um array de 2 dimensões

COm o exemplo a baixo, as localizações ficam = linha 1 e coluna 0 + linha 2 (3-1) e coluna 1 (2-1). Essas localizações dariam os números 13 e 10, contando os que fazem margem com ambos.

```python
arr[1:3, 0:2]
```

Ordenar (sorting) um array

ordena linha por linha

```python
np.sort(arr)
```

Quais são os eixos (axis)

axis 0 - eixo vertical

axis 1 - eixo horizontal

Agora podemos usar o sort para ordenar de acordo com o eixo desejado


Vertical

```python
np.sort(arr, axis=0)
```

Horizontal

```python
np.sort(arr, axis=1)
```

Crie um array que contenha paenas os tamanhos das espécies e ordene por ordem crescente

```python
especies = np.array([[747, 89, 33, 5],
                    [623, 123, 32, 13],
                    [501, 22, 49, 2],
                    [116, 101, 42, 10],
                    [297, 56, 69, 22],
                    [613, 64, 27, 7],
                    [295, 84, 29, 14],
                    [692, 105, 72, 16],
                    [229, 103, 35, 5],
                    [374, 124, 70, 1]])
especies
```

Selecione a segunda coluna com a quantidade de espécies encontradas e adicione em um arrey como 'qtd_especies'

```python
qtd_especies = especies[:, 1]
qtd_especies
```

De qtd_especies selecione apenas as primeiras 3 quantidades e print

```python
print(qtd_especies[0:3])
```

Print as 5 últimas quantidades de espécies

```python
print(qtd_especies[5:10])
```

  Também pode ser feito da seguinte forma, gerando um array:

  Essa técnica ajuda muito quando temos uma quantidade imensa de dados, pois vai direto para os últimos dados, sem qeu você precise ficar contabilizando linhas e quantidade de elementos.


```python
qtd_especies[-5:]
```

Crie um array que contenha apenas os tamanhos das espécies e ordene por ordem crescente

O tamanho das espécies é a coluna 4, sendo reconhecida como 3.

```python
np.sort(especies[:, 3])
```

Crie um array que contenha apenas os tamanhos das espécies e ordene, agora, por ordem decrescente

O tamanho das espécies é a coluna 4, sendo reconhecida como 3.

[::-1] - Array em ordem decrescente no range de python

```python
np.sort(especies[:, 3])[::-1]
```
