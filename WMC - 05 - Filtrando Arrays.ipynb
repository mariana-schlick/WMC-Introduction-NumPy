Filtrando arrays

Quando queremos filtrar os dados com base em condições que eles apresentam.

Indexação sofisticada (fancy indexing) ou usando máscara (mask) em array 2d

Exemplo: Queremos descobrir quais valores são divisíveis por 2.

Recebemos uma resposta em booleanos

```python
pessoas_id_idade = np.array([[1, 22], [2, 21], [3, 27], [4, 26]])
pessoas_id_idade[:, 1] % 2 == 0
```

Retornar pessoas com idade maior que 21

```python
mask_maior21 = pessoas_id_idade[:, 1] > 21
pessoas_id_idade[mask_maior21]
```

Exemplo com mask - retorna booleano de true ou false.

```python
arr1 = np.array([1, 2, 3, 4, 5])
mask = arr1 % 2 == 0
mask
```

Exemplo com fancy indexing - Retorna valores do array quando passado a máscara

```python
arr1[mask]
```

np.where() - retorna um array de indices e cria uma matriz com base nas condições dadas e os elementos que correspondem a estas condições.

No exemplo a baixo, ele retornou [0, 3] que correspondem ao 22 e ao 26.



```python
pessoas_id_idade = np.array([[1, 22], [2, 21], [3, 27], [4, 26]])
np.where(pessoas_id_idade[:, 1] % 2 == 0)
```

np.where com fancy indexing - irá retornar a localização do elemento na matriz.

Pelo exemplo abaixo, ele retornou linhas [0, 0, 1, 2, 2, 3] e depois colunas [2, 3, 3, 0, 2, 2], que na ordem simultaneamente, indicam um elemento.

[0, 2] = linha 0 e coluna 2 = 3

[0, 3] = linha 0 e coluna 3 = 15

[1, 3] = linha 1 e coluna 3 = 27

etc

```python
cartela_bingo = np.array([[16, 10, 3, 15],
                       [14, 23, 17, 27],
                       [6, 19, 3 , 1],
                       [10, 4, 18, 19]])
cartela_bingo
```

```python
np.where(cartela_bingo % 3 == 0)
```

Buscar e Substituir

No exemplo abaixo ele retornou o dtype'<U21' revelando que o array foi transformado em string. Onde se encontravam números divisíveis por 3 no array, será substituído por xxx.

```python
np.where(cartela_bingo % 3 == 0, "xxx", cartela_bingo)
```

Usando um index booleano, crie um array que contém os dados da maior espécie encontrada (considerando seu tamanho), esse valor corresponde ao valor 22.

Usando fancy index faça um array que contém apenas dados da espécie com ID 297.

Considere a profundeza em que a espécie foi encontttrada e substitua valores maiores que 60 por "Profundo"

```python
especies = np.array([[747, 89, 33, 5],
                    [623, 123, 32, 13],
                    [501, 22, 49, 2],
                    [116, 101, 42, 10],
                    [297, 56, 69, 22],
                    [613, 64, 27, 7],
                    [295, 84, 29, 14],
                    [692, 105, 72, 16],
                    [229, 103, 35, 5],
                    [374, 124, 70, 1]])
especies
```

Usando um index booleano, crie um array que contém os dados da maior espécie encontrada (considerando seu tamanho), esse valor corresponde ao valor 22.

```python
especies[:, 3] == 22
```

```python
maior_especie = especies[especies[:, 3] == 22]
maior_especie
```

Usando fancy index faça um array que contém apenas dados da espécie com ID 297.

```python
especies[:, 0] == 297
```

```python
especie297 = especies[especies[:, 0] == 297]
especie297
```

Usando np.where() faça um array com a linha com dados correspondentes a espécie com 105 representantes encontrados.

```python
np.where(especies[:, 1] == 105)
```

```python
especies[np.where(especies[:, 1] == 105)]
```
